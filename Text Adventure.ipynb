```python
print(""" 
     _      __    __                     __                                 
    | | /| / /__ / /______  __ _  ___   / /____                             
    | |/ |/ / -_) / __/ _ \/  ' \/ -_) / __/ _ \                            
    |__/|__/\__/_/\__/\___/_/_/_/\__/__\__/\___/ __  ________  ____  _  __  
                  / __/ __/ __/  \ \/ / __ \/ / / / / __/ __ \/ __ \/ |/ /  
                 _\ \/ _// _/     \  / /_/ / /_/ / _\ \/ /_/ / /_/ /    / _ 
                /___/___/___/     /_/\____/\____/ /___/\____/\____/_/|_/ (_)
                                                                            
                                                                                                                                                                                                                                                                                                                                        
""")

#Importing Random package.
import random

#Defining of Game_Start function for getting user inputs
def Game_Start():
    
    #This function starts the game and gets the input from user.
    
    #Displaying the messages to user by using print function.
    print("Do you want to Start the Game ??\n\n")
    print("PRESS Y to START THE GAME <----------> PRESS N FOR EXIT\n\n")
    
    #Assigning value to start_var variable by using input fuction.
    start_var = input(prompt = ">>>>>")
    
    #Converting user input to lower case by using lower()
    #using strip() to obtain the trimmed version of the input.
    check_case = start_var.strip().lower()
    
    #Conditional statement for checking the user input by if statement.
    if check_case =='y' or check_case == 'yes':
        #print statements to display message to user.
        print("THE GAME BEGINS...... GOOD LUCK\n\n\n")
        
        print( """You need to Meet Undercover agent to collect the details of kidnapped girl
        
         Press 1 to meet the Agent
         Press 2 to Fail\n\n""")
        
        #Declaring and assigning value to local variable user_input.
        user_input = input(prompt =  ">>>>>")
        
        # Start of nested conditional statements
        #Checking user_input variable value using if condition.
        if user_input == '1':
            print("That's Awesome ---> Going to Meet Undercover Agent.\n\n\n")
            
            #Invoking Meet_Undercover_Agent().
            Meet_Undercover_Agent()
 
           
        #Checking user_input variable value using elif condition.   
        elif user_input =='2':
            print("Sorry You have Failed. Thank you for playing !!\n\n\n")
            input(prompt = '<Press any key to continue>\n\n')
            
            #Invoking fail().
            fail()
            
        #Default else condition if the above nested conditional statements fail. 
        else:
            print("Entered value doesn't meets the criteria .")
            print("Do you want to try again")
            
            #Invoking Game_Start() for retrying.
            Game_Start()
            
        #End of nested conditional statements.
        
    #Conditional statement for checking the user input by elif statement    
    elif check_case == 'n' or check_case == 'no':
        print("Ohh Sad !! See You Later\n\n")
        #Invoking fail().
        fail()
        
    #Default else statement
    else:
        print("Invalid entry. Please try again.\n\n")
        #Getting user input 
        input(prompt = '<Press any key to continue>\n\n')
        
        #Invoking Game_Start() for retrying.
        Game_Start()
    

#Defining of Meet_Undercover_Agent function for proceeding to stage 2 based on user inputs.
def Meet_Undercover_Agent():
    #This function proceeds to stage 2 by checking user inputs.
    
    #Displaying message to user using print()
    print("Press 1. To Meet the Agent.  \t  Press 2. to Fail\n")
    
    #Assigning value to local variable user_input 
    user_input= input(prompt = "Enter 1 or 2\n\n")
    
    #checking user_input value by using if condition.
    if user_input=='1':
            print("""Undercover Agent shared the valuable information about kidnapped Woman's Locations.

       ------------GREAT JOB BUDDY-----------
       
       Proceed to Stage 2
              |
              v
              
       Find the Location by Using GPS Locations
       
       
          """)
            #Invoking get_Cordinate().
            get_Cordinate()
            #input("Press any key to search the location")
    #checking user_input value by using elif condition.        
    elif user_input=='2':
    #Invoking fail().    
            fail()
    #Default else statement if the above two conditions fail.       
    else:    
        print("Invalid Entry. Do you want to try Again\n")
        
        #Invoking Meet_Undercover_Agent() for retrying.
        Meet_Undercover_Agent()
        
        

#Defining of get_Cordinate function for proceeding to activate GPS and user inputs.
def get_Cordinate():
    #This function is used to activate the GPS based on the user selection
    
    #Assigning input value to cordinate_input variable
    cordinate_input = input(prompt="Enter YES to activate GPS and NO to FAIL\n\n")
    
    #Converting the value of cordinate_input variable to Capitalize format and assigning to new variable. 
    #Using strip() to trim the input string
    check_upper  = cordinate_input.strip().capitalize()
    
    #Checking user_input value by using if condition.
    if check_upper == 'Yes' or check_upper =='Y':
        #Displaying message to user by using print().
        print("GPS Activated , Searching for Location\n\n")
        #using input() for user input.
        input(prompt = "press any key to confirm the location\n\n")
        #Displaying messages to user by using print().
        print("Location Found\n\n")
        print("""You have reached the Enemy territory.
        
        Play a Guessing Game to Open the Door.
        
        Guess a number between 1 and 5 to Open the DOOR
        
        All the best !!! \n\n """)
        #Invoking Open_Door()to play the guessing game.
        Open_Door()
    #Checking user_input value by using elif condition.        
    elif check_upper == 'No'or check_upper == 'N':
        print("Sorry You have failed\n\n")
        #Invoking fail().
        fail()
    #Default else statement if the above two conditions fail.     
    else:
        print("Invalid Entry--- Do you want to try again")
        #Invoking get_Cordinate() to retry.
        get_Cordinate()
        
#Defining Open_Door() to Open the enemy territory door based on a guessing game.    
def Open_Door():
    #Open_Door function is used to play a guessing game based on random numbers.
    
    #Assigning a random value from 1 to 5 to the guess_value by using randint function.
    guess_value= random.randint(1,5)
    #Setting a counter variable to check the number of retries.
    tries=5
    #while loop
    while tries > 0:
        number_input= int(input(prompt = "Guess an integer from 1 to 5\n"))
        #Checking if the number_input is equal to guess_value.
        if number_input == guess_value:
            
            print("""Wonderful the number is correct
            THE DOOR IS OPEN NOW!!
            
            There are 5 Enemies to Attack you !!
            \n\n""")
            #Assigning zero to variable tries.
            tries=0
            #Invoking kill_enemy() to kill the enemies
            kill_enemy()
        #Checking if the number_input is not equal to guess_value.    
        elif number_input != guess_value:
            #Decrementing the value of tries using assignment function.
            tries = tries-1
            print(f"""Incorrect number , Remaining Guesses-------> {tries}
            Choose a Different Number\n\n\n""")
        #Default else statement.
        else:
            print("Better luck next time")
            #Invoking fail().
            fail()
            
#Defining kill_enemy() to kill the enemies and rescue the hostage.           
def kill_enemy():
    #Assigning the value five to enemy_count variable.
    enemy_count = 5
    #While loop to check the remaining enemy count.
    while enemy_count > 0:
        #Getting user input by using input().
        input(prompt = '<Press any key to kill the enemy>\n')
        #Decrementing enemy_count variable.
        enemy_count = enemy_count-1
        print(f"""Enemies remaining-----> {enemy_count},<---- Shoot Them """)
        #Checking if enemy_count is equal to zero.
        if enemy_count == 0:
            #Invoking win() when enemy_count is zero.
            win()
            
#Defining win() to display the winner details.
def win():
        
    print("YOU WON -- GREAT ACHIEVEMENT")
    print("""  _    _  __  _  _  _  _  ___  ___  
( \/\/ )(  )( \( )( \( )(  _)(  ,) 
 \    /  )(  )  (  )  (  ) _) )  \ 
  \/\/  (__)(_)\_)(_)\_)(___)(_)\_) """)

#Defining fail() to exit the game when the user fails on any of the stages.
def fail():
        
    print("You've failed, but thanks for playing!")
    input(prompt = '<Press any key to exit>\n')

#Invoking Game_Start fuction to start the game.
Game_Start()


"""GAME REFERENCES ----> 1. Patorjk.com"""
```
